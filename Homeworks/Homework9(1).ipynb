# Dimension Reduction techniques

Dimension Reduction is nothing but converting the dimensions of huge data into a lesser data dimensions. And these are useful in Machine Learning and Predictive Analysis. Let us discuss with an example : The below figure shows the 2D plot between two co-ordinates x2(inches) and x1(cm) which has some noise in it. But when it is converted into a 1D plot (z1) can be easily understood with low noise. In similar ways, we can reduce n dimensions of data set to k dimensions (k < n) . These k dimensions can be directly identified (filtered) or can be a combination of dimensions (weighted averages of dimensions) or new dimension(s) that represent existing multiple dimensions well.

For example, you have an image containing more number of pixels and to reduce them you can't just randomly eliminate the pixels and for that to know which fields need to get eliminated in a data we require these dimension reduction techniques.

# Benefits of Dimension Reduction

* As the data is being compressed it reduces the storage space.
* Decreases the time consumption to analyse.
* Can be able to test with different algorithms as it consists of lesser dimensions.
* It eliminates redundant features.(like storing the same field with different units).
* Helps in Noise elimination.

# Common methods to perform Dimension Reduction :

*MISSING VALUES :
    In the data exploration if we encounter the problem of missing fields we need to perform two either imputing the missing values or to drop the filed itself. Imputation is the process in which the missing values are replaced with the substitutional values. If the field contains 40% of missing values you can drop the field completely.

*LOW VARIANCE :
    The variation in the value of a field describes to analyse how good or bad the impact is on prediction. But if the variable is a constant value there is no use in analysing the high dimensional data, and hence these variables are considered to have zero variance and can be dropped when compared to others.

*DECISION TREES:
     These are the effectively used to handle many challenges like identifying significant variables,outliers and missing values.Even used by several people at a data science compitetion namely "Data Hackathon".

*PRINCIPAL COMPONENT ANALYSIS(PCA):
    Two important aspect in data analysis is data exploration and representation. The higher dimensional data can be defined in a two-dimensional space data using the Principal Component Analysis (PCA) along with the Analysis. On the other hand we can find the hidden relationships in data by grouping them into clusters.
     PCA reduces the dimensionality by finding the principal components. The dataset is changed into a reduced dataset based on these principal components. And this reduced dataset can be represented in a two-dimensional chart.(Example : Scatterplot).

# PRINCIPLE COMPONENT ANALYSIS :

PCA is an unsupervised technique, which means that information about the class labels is ignored.The variables are converted into new set of variables which are linear combination of original variables and are called as principal components. For example for a two dimensional dataset there would be two principle components and these components are orthogonal to each other as shown below.

INITIAL SETUP OF DATA

```python
%pylab inline
import pandas as pd
df_wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data")
df_wine
```

Eigenvalues & eigenvectors

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
X, y = df_wine.iloc[: , 1 : ] . values, df_wine.iloc [ : , 0].values
X_train,  X_test, y_train, y_test = \
         train_test_split(X, y, test_size =0.3, random_state=0)
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
```

```python
covariant_matrix = np.cov(X_train_std.T)
covariant_matrix[0 :: 5]
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariant_matrix)
eigen_values, eigen_vectors[ : : 5 ]
```

```python
tot = sum(eigen_values)
var_exp = [(i / tot) for i in sorted(eigen_values, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
bar(range(1,14), var_exp, alpha=0.5, align= 'center', label = 'individual explained variance') 
step(range(1,14), cum_var_exp, where= 'mid' , label = 'cumulative explained variance')
ylabel('Explained variance ratio')
xlabel('Principal components')
legend(loc= 'best')
show()
```

The plot shows that the first principal component alone accounts for 40 percent of the variance. Also, we can see that the first two principal components combined explain almost 60 percent of the variance in the data.



# Transform dataset onto a new principal axes

```python
eigen_pairs = \
[(np.abs(eigen_values [i]), eigen_vectors[: , i]) for i in range (len(eigen_values))]
eigen_pairs.sort(reverse=True)
```

# Classification of transformed samples using logistic regression.

```python
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
# setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
lr = LogisticRegression()
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
lr.fit(X_train_pca, y_train)
plot_decision_regions(X_train_pca, y_train, classifier = lr)
xlabel('PC1')
ylabel('PC2')
legend(loc = 'upper right')
show()
```

Decision regions for the training model are reduced to the two principal component axes.
